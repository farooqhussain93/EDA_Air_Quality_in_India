```python
def save_plot(name):
    plt.savefig(fr"D:\Farooq_Hussain\Python\Data Analysis\04 - EDA_Global_Air_Quality\images/{name}.png", dpi=300, bbox_inches="tight")
```

```python
# Q1. Are there any missing data patterns in the dataset?
# Heatmap of missing values

plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
save_plot('Missing_Values_Heatmap')
plt.show()
```

```python
# Bar plot of % missing per column

missing_percent = df.isnull().mean() * 100

plt.figure(figsize=(10,5))
missing_percent.sort_values(ascending=False).plot(kind='bar', color='tomato')
plt.ylabel('% Missing')
plt.title('Percentage of Missing Values per Column')
save_plot('Percentage_of_Missing_Values_per_Column')
plt.show()
```

```python
# Now I am going to clean this dataset

pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 
              'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']

# Dropping rows where AQI and all pollutant values are missing
df = df.dropna(subset=['AQI'] + pollutants, how='all')
df = df.dropna(subset=['AQI'])

# Filling missing pollutant values with median per city
for col in pollutants:
    df[col] = df.groupby('City')[col].transform(lambda x: x.fillna(x.median()))
    
df.isnull().sum()
```

```python
# Q2. Which cities have the highest average AQI (Air Quality Index)?

# Group by city and calculate mean AQI (descending order)
city_aqi_avg = df.groupby('City')['AQI'].mean().sort_values(ascending=False)

print(f"Cities with highest average AQI:\n\n{city_aqi_avg.head(10)}")
```

```python
# Q3. Bar chart of Top 10 most polluted cities by AQI.

plt.figure(figsize=(10,6))
sns.barplot(
    x=city_aqi_avg.head(10).values, 
    y=city_aqi_avg.head(10).index, 
    hue=city_aqi_avg.head(10).index,  # using hue explicitly
    palette='Reds_r', 
    legend=False  # hide legend since it's redundant
)
plt.title("Top 10 Most Polluted Cities by Average AQI")
plt.xlabel("Average AQI")
plt.ylabel("City")
save_plot('Top_10_Most_Polluted_Cities_by_Average_AQI')
plt.show()
```

```python
# Q4. Which city has the cleanest air on average?

# Group by city and calculate mean AQI (ascending order)
city_AQI_avg = df.groupby('City')['AQI'].mean().sort_values()

print(f"Cities with lowest average AQI:\n\n{city_AQI_avg.head(10)}")
```

```python
# Plot of cleanest cities (lowest average AQI)

plt.figure(figsize=(10,6))
sns.barplot(
    x=city_AQI_avg.head(10).values, 
    y=city_AQI_avg.head(10).index, 
    hue=city_AQI_avg.head(10).index, 
    palette='Greens', 
    legend=False
)
plt.title("Top 10 Cleanest Cities by Average AQI")
plt.xlabel("Average AQI")
plt.ylabel("City")
save_plot('Top_10_Cleanest_Cities_by_Average_AQI')
plt.show()
```

```python
# Q5. How does AQI vary by month or season?

# Ensure Date column is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and season
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.strftime('%b')

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['Season'] = df['Month'].apply(get_season)


monthly_aqi = df.groupby('Month_Name')['AQI'].mean().reindex(
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
)

plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_aqi.index, y=monthly_aqi.values, marker='o', color='orange')
plt.title("Average AQI by Month")
plt.xlabel("Month")
plt.ylabel("Average AQI")
save_plot('Average_AQI_by_Month')
plt.show()
```

```python
# Q6. Seasonal comparison plot of average AQI levels.

seasonal_aqi = df.groupby('Season')['AQI'].mean().reindex(['Winter','Spring','Summer','Autumn'])

plt.figure(figsize=(8,5))
sns.barplot(x=seasonal_aqi.index, y=seasonal_aqi.values, hue=seasonal_aqi.index, palette='coolwarm')
plt.title("Average AQI by Season")
plt.xlabel("Season")
plt.ylabel("Average AQI")
save_plot('Average_AQI_by_Season')
plt.show()
```

```python
# Q7. Map visualization showing geographic distribution of AQI values.

unique_cities = df["City"].unique()
print(len(unique_cities), "cities found")
```

```python
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="aqi_project")

city_coords = {}

for city in unique_cities:
    try:
        location = geolocator.geocode(city + ", India")  # restrict to India
        if location:
            city_coords[city] = [location.latitude, location.longitude]
        else:
            city_coords[city] = None
        time.sleep(1)  # sleep to avoid hitting API too fast
    except:
        city_coords[city] = None

print(city_coords)
```

```python
city_aqi = df.groupby("City")["AQI"].mean().reset_index()
city_aqi["coords"] = city_aqi["City"].map(city_coords)
city_aqi = city_aqi.dropna(subset=["coords"])
```

```python
import folium

m = folium.Map(location=[22, 78], zoom_start=5)

for _, row in city_aqi.iterrows():
    folium.CircleMarker(
        location=row["coords"],
        radius=7,
        popup=f"{row['City']}: {row['AQI']:.2f}",
        color="red" if row["AQI"] > 150 else "green",
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

m.save("aqi_map.html")
```

```python
# Q8. Boxplot comparing AQI across major cities.

top_cities = df['City'].value_counts().head(10).index
df_top = df[df['City'].isin(top_cities)]

plt.figure(figsize=(12,6))
sns.boxplot(x='City', y='AQI', data=df_top, hue='City', palette='Set2', legend=False)
plt.title("Boxplot Comparing AQI Across Major Cities")
plt.xlabel("City")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
save_plot('Boxplot_Comparing_AQI_Across_Major_Cities')
plt.show()
```

```python
# Q9. What pollutants are recorded, and what are their average levels?

pollutants = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']


pollutant_avg = df[pollutants].mean().sort_values(ascending=False)

print(pollutant_avg)
```

```python
# Q10. Heatmap showing correlation between pollutants (PM2.5, PM10, NO2, SO2, O3, CO).

main_pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'O3', 'CO']

# Correlation Matrix
corr = df[main_pollutants].corr()

# Heatmap Plot
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Major Pollutants")
save_plot('Correlation_Between_Major_Pollutants')
plt.show()
```

```python
# Q11. How do pollution levels vary between weekdays and weekends?

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df['DayOfWeek'] = df['Date'].dt.dayofweek
df['DayType'] = df['DayOfWeek'].apply(lambda x: 'Weekend' if x > 5 else 'Weekday')

daytype_aqi = df.groupby('DayType')['AQI'].mean()

plt.figure(figsize=(6,5))
sns.barplot(x=daytype_aqi.index, y=daytype_aqi.values, hue=daytype_aqi.index, palette="Set2", legend=False)
plt.title("Average AQI: Weekdays vs Weekends")
plt.ylabel("Average AQI")
plt.xlabel("")
save_plot('Average_AQI_Weekdays_vs_Weekends')
plt.show()
```

```python
# Q12. Stacked bar charts of pollutant contributions in different cities.

pollutants = ["PM2.5", "PM10", "NO2", "SO2", "O3", "CO"]

# Calculating average pollutant levels per city
city_pollutants = df.groupby("City")[pollutants].mean()

# Normalizing each row so contributions sum to 1 (percentage share)
city_pollutants_norm = city_pollutants.div(city_pollutants.sum(axis=1), axis=0)

# Top 10 polluted cities by AQI
top10_cities = df.groupby("City")["AQI"].mean().sort_values(ascending=False).head(10).index
city_pollutants_top10 = city_pollutants_norm.loc[top10_cities]


city_pollutants_top10.plot(
    kind="bar",
    stacked=True,
    figsize=(12,7),
    colormap="tab20"
)

plt.title("Pollutant Contributions in Top 10 Most Polluted Cities")
plt.ylabel("Proportion of Pollution")
plt.xlabel("City")
plt.legend(title="Pollutant", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
save_plot('Pollutant_Contributions_in_Top_10_Most_Polluted_Cities')
plt.show()
```

```python
# Q13. Which cities consistently exceed safe pollution limits?

# In India, according to CPCB standards,

# AQI ≤ 100 → “Good / Satisfactory” (safe)

# AQI > 100 → “Moderate / Poor / Very Poor / Severe” (unsafe)

# Defining safe threshold
safe_limit = 100

# A new column marking if AQI exceeded safe limit
df["Above_Safe_Limit"] = df["AQI"] > safe_limit

# Calculate percentage of unsafe days per city
city_exceed = (
    df.groupby("City")["Above_Safe_Limit"]
    .mean()
    .sort_values(ascending=False) * 100
).reset_index()


# Top 10 worst offenders
print(city_exceed.head(10))

# Visualization
plt.figure(figsize=(12,6))
sns.barplot(
    data=city_exceed.head(10),
    x="City", y="Above_Safe_Limit", hue="City", legend=False,
    palette="Reds_r"
)

plt.xticks(rotation=45)
plt.ylabel("% Days AQI > 100")
plt.title("Cities Consistently Exceeding Safe Pollution Limits")
save_plot('Cities_Consistently_Exceeding_Safe_Pollution_Limits')
plt.show()
```
